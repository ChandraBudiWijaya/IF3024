```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Membuat array waktu t
T = np.arange(0, 2, 0.0001)

# 2. Membuat sinyal-sinyal sesuai tugas
y1 = 2 * np.sin(2 * np.pi * 3 * T + 0)
y2 = 1 * np.cos(2 * np.pi * 4 * T + np.pi/4)
y3 = -1 * np.sin(2 * np.pi * 5 * T + np.pi/2)
y4 = 0.5 * np.cos(2 * np.pi * 6 * T + np.pi)
```

```python
# 3. Membuat subplot untuk membandingkan sinyal
fig, axs = plt.subplots(4, 1, figsize=(10, 12))
axs[0].plot(T, y1, label='y1')
axs[0].set_title('y1 = 2 * sin(2π * 3 * t)')
axs[1].plot(T, y2, label='y2', color='r')
axs[1].set_title('y2 = cos(2π * 4 * t + π/4)')
axs[2].plot(T, y3, label='y3', color='g')
axs[2].set_title('y3 = -sin(2π * 5 * t + π/2)')
axs[3].plot(T, y4, label='y4', color='m')
axs[3].set_title('y4 = 0.5 * cos(2π * 6 * t + π)')

for ax in axs:
    ax.legend()
    ax.set_xlabel('Time (t)')
    ax.set_ylabel('Amplitude')
    ax.grid()

plt.tight_layout()
plt.show()
```

```python
# 4. Membuat sinyal kombinasi y5
y5 = y1 + y2

# 5. Plot y1, y2, and y5 with enhanced precision
plt.figure(figsize=(15, 8))
plt.plot(T, y1, label='y1 = 2*sin(2π*3*t)', color='b', linewidth=1.5)
plt.plot(T, y2, label='y2 = cos(2π*4*t + π/4)', color='r', linewidth=1.5)
plt.plot(T, y5, label='y5 = y1 + y2', color='k', linestyle='--', linewidth=2)

plt.title('Sinyal y1, y2, dan Kombinasinya (y5)', fontsize=12, pad=15)
plt.xlabel('Time (t)', fontsize=10)
plt.ylabel('Amplitude', fontsize=10)

# Enhance grid
plt.grid(True, linestyle='--', alpha=0.7)

# Set axis limits for better visibility
plt.xlim(0, 1)  # Show first second of the signal

# Add legend
plt.legend()

# Show plot
plt.show()
```
