```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Membuat array waktu t dari 0 hingga 2 dengan langkah 0.0001
T = np.arange(0, 2, 0.0001)
```

```python
# 2. Membuat sinyal
y1 = 2 * np.sin(2 * np.pi * 3 * T + 0)  # Sinyal sinus dengan amplitudo 2 dan frekuensi 3 Hz
y2 = 1 * np.cos(2 * np.pi * 4 * T + np.pi/4)  # Sinyal cosinus dengan amplitudo 1, frekuensi 4 Hz, dan pergeseran fase π/4
y3 = -1 * np.sin(2 * np.pi * 5 * T + np.pi/2)  # Sinyal sinus negatif dengan amplitudo 1, frekuensi 5 Hz, dan pergeseran fase π/2
y4 = 0.5 * np.cos(2 * np.pi * 6 * T + np.pi)  # Sinyal cosinus dengan amplitudo 0.5, frekuensi 6 Hz, dan pergeseran fase π
```

```python
# Membuat subplot 2x2 dengan ukuran gambar 10x6 inci
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# Plot sinyal y1 di subplot pertama (kiri atas)
axs[0, 0].plot(T, y1, label='y1')
axs[0, 0].set_title('y1 = 2 * sin(2π * 3 * t)')  # Menambahkan judul untuk y1

# Plot sinyal y2 di subplot kedua (kanan atas) dengan warna merah
axs[0, 1].plot(T, y2, label='y2', color='r')
axs[0, 1].set_title('y2 = cos(2π * 4 * t + π/4)')  # Menambahkan judul untuk y2

# Plot sinyal y3 di subplot ketiga (kiri bawah) dengan warna hijau
axs[1, 0].plot(T, y3, label='y3', color='g')
axs[1, 0].set_title('y3 = -sin(2π * 5 * t + π/2)')  # Menambahkan judul untuk y3

# Plot sinyal y4 di subplot keempat (kanan bawah) dengan warna magenta
axs[1, 1].plot(T, y4, label='y4', color='m')
axs[1, 1].set_title('y4 = 0.5 * cos(2π * 6 * t + π)')  # Menambahkan judul untuk y4

# Melakukan iterasi untuk setiap subplot untuk menambahkan elemen umum
for ax in axs.flat:
    ax.legend()  # Menambahkan legenda ke setiap subplot
    ax.set_xlabel('Waktu (t)')  # Menambahkan label di sumbu x
    ax.set_ylabel('Amplitudo')  # Menambahkan label di sumbu y
    ax.grid()  # Menambahkan grid ke setiap subplot

# Mengatur jarak antar subplot
plt.tight_layout()

# Menampilkan seluruh gambar
plt.show()
```

```python
# 4. Membuat sinyal y5 kombinasi dari y1 dan y2
y5 = y1 + y2

# 5. Plot y1, y2, dan y5
plt.figure(figsize=(15, 8))
plt.plot(T, y1, label='y1 = 2*sin(2π*3*t)', color='b', linewidth=1.5)
plt.plot(T, y2, label='y2 = cos(2π*4*t + π/4)', color='r', linewidth=1.5)
plt.plot(T, y5, label='y5 = y1 + y2', color='k', linestyle='--', linewidth=2)

plt.title('Sinyal y1, y2, dan Kombinasinya (y5)', fontsize=12, pad=15)
plt.xlabel('Time (t)', fontsize=10)
plt.ylabel('Amplitude', fontsize=10)

# Menambahkan grid
plt.grid(True, linestyle='--', alpha=0.7)

# Mengatur batas sumbu untuk visibilitas yang lebih baik
plt.xlim(0, 1)  # Menampilkan satu detik pertama dari sinyal

# Menambahkan legenda
plt.legend()

# Menampilkan plot
plt.show()
```
